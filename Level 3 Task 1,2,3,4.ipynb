### Level 3 Task 1
 Task 1: Restaurant Reviews
 Analyze the text reviews to identify the most
 common positive and negative keywords.
 Calculate the average length of reviews and
 explore if there is a relationship between
 review length and rating

```python
import pandas as pd
from collections import Counter
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Task: Analyze text reviews to identify positive and negative keywords
data = pd.read_csv("Dataset.csv")
reviews = data["Rating text"]
ratings = data["Aggregate rating"]

sia = SentimentIntensityAnalyzer()
positive = []
negative = []

for review in reviews:
    words = review.lower().split()
    score = sia.polarity_scores(review)["compound"]
    if score > 0:
        positive.extend(words)
    elif score < 0:
        negative.extend(words)

positive_keywords = Counter(positive).most_common(10)
negative_keywords = Counter(negative).most_common(10)

# Task: Calculate the average length of reviews
average_length = reviews.str.len().mean()

# Task: Explore relationship between review length and rating
correlation = reviews.str.len().corr(ratings)

print("Positive Keywords:", positive_keywords)
print("Negative Keywords:", negative_keywords)
print("Average Review Length:", average_length)
print("Correlation between Review Length and Rating:", correlation)
```

```python
# Calculate the average length of reviews and
#  explore if there is a relationship between
#  review length and rating
```

```python

review_lengths = reviews.apply(len)
average_length = review_lengths.mean()

# Task: Explore the relationship between review length and rating
correlation = review_lengths.corr(ratings)

print("Positive Keywords:", positive_keywords)
print("Negative Keywords:", negative_keywords)
print("Average Review Length:", average_length)
print("Correlation between Review Length and Rating:", correlation)
```

### Level 3 Task 2
> Task: Votes Analysis

```python
#  Identify the restaurants with the highest and lowest number of votes
 
```

```python
import pandas as pd

# Load the dataset
data = pd.read_csv("Dataset.csv")

# Find the restaurant with the highest number of votes
highest_votes = data.loc[data["Votes"].idxmax()]

# Find the restaurant with the lowest number of votes
lowest_votes = data.loc[data["Votes"].idxmin()]

print("Restaurant with the Highest Votes:")
print(highest_votes)

print("\nRestaurant with the Lowest Votes:")
print(lowest_votes)
```

### Level 3 Task 3
> Task: Price Range vs. Online Delivery and  Table Booking


```python
 # Analyze if there is a relationship between the  price range and the availability of online  delivery and table booking.

```

```python
import pandas as pd

# Load the dataset
data = pd.read_csv("Dataset.csv")

# Group data by Price Range and analyze Online Delivery and Table Booking
analysis = data.groupby("Price range").agg({
    "Has Online delivery": "value_counts",
    "Has Table booking": "value_counts"
})

print("Relationship between Price Range and Availability:")
print(analysis)
```

```python
 # Determine if higher-priced restaurants are more likely to offer these services
```

```python
import pandas as pd

# Load the dataset
data = pd.read_csv("Dataset.csv")

# Analyze the proportion of restaurants offering services by price range
service_analysis = data.groupby("Price range").agg({
    "Has Online delivery": lambda x: (x == "Yes").mean(),
    "Has Table booking": lambda x: (x == "Yes").mean()
})

print("Proportion of Services Offered by Price Range:")
print(service_analysis)
```
